```python
import os
import fastf1
import pandas as pd
import plotly.express as px

os.chdir('C:/Users/user/OneDrive/Desktop/MyProjects/Project2')
fastf1.Cache.enable_cache('cache')

print('Setup complete!')
```

```python
def race_process(year, race_name):
    session = fastf1.get_session(year, race_name, 'R')
    session.load() #loaded the session

    racelaps = session.laps.copy() #copied all the laps for manipulation

    racelaps['LapTimeSeconds'] = racelaps['LapTime'].dt.total_seconds() #created a new column with lap time in seconds for easier analysis

    #filtering all the laps based on rules and conditions
    cleaned_laps = racelaps.dropna(subset=['LapTimeSeconds']) #dropped laps with no lap time (e.g. pit in/out, out laps)
    cleaned_laps = cleaned_laps[cleaned_laps['PitInTime'].isna()] #dropped laps where the driver pitted in
    cleaned_laps = cleaned_laps[cleaned_laps['PitOutTime'].isna()] #dropped laps where the driver pitted out
    cleaned_laps = cleaned_laps[cleaned_laps['Compound'].isin(['SOFT', 'MEDIUM', 'HARD'])] #kept only laps on the main dry compounds
    
    def remove_outliers(driver_laps):
        median = driver_laps['LapTimeSeconds'].median()
        return driver_laps[(driver_laps['LapTimeSeconds'] <= 1.10 * median)] #removing laps that are more than 10% slower than the median lap time for that driver
    
    cleaned_laps = cleaned_laps.groupby('Driver', group_keys = False).apply(remove_outliers) #applied the outlier removal function to each driver's laps

    #saving the cleaned data to a csv file for future use
    cleaned_laps.to_csv(f"data/{race_name.lower().replace(' ', '_')}_{year}_clean.csv", index = False)

    #generating the charts for the cleaned data
    fig1 = px.scatter(cleaned_laps, x = 'LapNumber', y = 'LapTimeSeconds', color = 'Driver', title = f'Lap Time Evolution - {year} {race_name}')
    fig1.write_html(f'charts/lap_times_{race_name.lower().replace(" ", "_")}_{year}.html')  #scatter plot of lap times over the course of the race generated and saved as an HTML file

    fig2 = px.box(cleaned_laps, x = 'Compound', y = 'LapTimeSeconds', color = 'Driver', title = f'Lap Time Distribution by Tyre Compound - {year} {race_name}')
    fig2.write_html(f'charts/compound_analysis_{race_name.lower().replace(" ", "_")}_{year}.html') #box plot of lap time distribution by tyre compound generated and saved as an HTML file  
```

```python
races = ['Japan', 'Bahrain', 'Saudi Arabia', 'Australia', 'China']
for i in races:
    try:
       race_process(2024, i)
       print(f'Successfully processed {i} GP!')
    except Exception as e:
        print(f'Error processing {i} GP: {e}')
```

```python
races = ['Japan', 'Bahrain', 'Saudi Arabia', 'Australia', 'China']
cols = ['Driver', 'Team', 'LapNumber', 'LapTimeSeconds', 'Compound', 'TyreLife', 'Stint', 'Position']

all_races = []
for race in races:
    filename = f"data/{race.lower().replace(' ', '_')}_2024_clean.csv"
    temp = pd.read_csv(filename)
    temp = temp[cols].copy()
    temp['Race'] = race
    all_races.append(temp)

#rewrote the cleaned data file so that it also contains the "Race" column for future use
```

```python
master_df = pd.concat(all_races, ignore_index=True)
print(master_df.shape)
print(master_df['Race'].value_counts())
```

```python
master_df.to_csv('data/master_clean.csv', index=False)
print("Master dataset saved!")
```
