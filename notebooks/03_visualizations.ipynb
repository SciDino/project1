```python
import os
import pandas as pd
import plotly.express as px

os.chdir('C:/Users/user/OneDrive/Desktop/MyProjects/Project2')

df = pd.read_csv('data/japan_2024_clean.csv')
print(df.shape)
```

```python
fig = px.scatter(df, x = 'LapNumber', y = 'LapTimeSeconds', color = 'Driver', title = 'Lap Time Evolution - 2024 Japanese GP')
fig.show()
```

```python
fig.write_html('charts/lap_times_japan_2024.html')
print('chart saved!')
```

```python
fig2 = px.box(df, x = 'Compound', y = 'LapTimeSeconds', color = 'Driver', title = 'Lap Time Distribution by Tyre Compound - 2024 Japanese GP')
fig2.show() 
```

```python
fig2.write_html('charts/compound_analysis_japan_2024.html')
print('chart saved!')
```
