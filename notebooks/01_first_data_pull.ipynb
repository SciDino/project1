```python
import fastf1
import pandas as pd
fastf1.Cache.enable_cache('C:/Users/user/OneDrive/Desktop/MyProjects/Project2/cache')  # Enable caching to speed up data retrieval
print("Fast F1 loaded successfully")
```

```python
session = fastf1.get_session(2024, 'Japan', 'R') #Get the session data for the 2024 Suzuka Grand Prix
session.load() #download the data for the session
print('Session loaded:', session.event['EventName']) #Print the name of the event to confirm data is loaded
```

```python
laps = session.laps #saves the lap data from the session into a variable called laps
print(laps)
print("Total laps:", len(laps))
print("Columns:", laps.columns.tolist()) #converting all the column names into a list for readability
```

```python
laps['LapTimeSeconds'] = laps['LapTime'].dt.total_seconds()
ver_laps = laps[laps['Driver'] == 'VER'].copy()
print(laps['LapTimeSeconds'])
print(ver_laps[['LapNumber', 'LapTime', 'LapTimeSeconds', 'Compound', 'TyreLife']])    
```

```python
fastest_lap = ver_laps['LapTimeSeconds'].min() #finding fastest lap time in seconds for Verstappen
print("Verstappen's fastest lap:", fastest_lap)
fastest_compound = ver_laps[ver_laps['LapTimeSeconds'] == fastest_lap]['Compound'].values[0] #finding the compound used for the fastest lap
print("Compound used for fastest lap:", fastest_compound)
```

```python
ver_laps_sorted = ver_laps.sort_values('LapTimeSeconds') #sorting Verstappen's laps by lap time and printing the five fastest ones
print(ver_laps_sorted[['LapTimeSeconds', 'Compound', 'TyreLife']].head(5)) #printing the 5 fastest lap times for Verstappen
```
